```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python

# File paths for population density and health facilities datasets
population_file_path = r"C:\Users\t_ongep\Downloads\Dalberg\Distribution-of-population-by-population-density-and-sub-county-2019.xlsx"
facilities_file_path = r"C:\Users\t_ongep\Downloads\Dalberg\ehealth-kenya-facilities-download.xlsx"

# Read population density and health facilities datasets
population_df = pd.read_excel(population_file_path)
facilities_df = pd.read_excel(facilities_file_path)

# Standardize Sub-County names in facilities_df to uppercase
facilities_df['Sub-County'] = facilities_df['Sub-County'].str.upper()
population_df['Sub-County'] = population_df['Sub-County'].str.upper()

population_df.head()
facilities_df.head()

# Replace 'Nairobi City'population_df  with 'Nairobi' in 'County' column so that the names are the same for merging
#population_df['County'] =population_df['County'].replace('Nairobi City', 'Nairobi')


# Define corrections dictionary for sub-county names and their population densities
corrections = {
    'DAGORETTI SOUTH': 'DAGORETTI',
    'DAGORETTI NORTH': 'DAGORETTI',
    'EMBAKASI EAST': 'EMBAKASI',
    'EMBAKASI WEST': 'EMBAKASI',
    'EMBAKASI SOUTH': 'EMBAKASI',
    'ROYSAMBU': 'EMBAKASI',  # Assuming Roysambu is part of Embakasi based on correction
    'EMBAKASI CENTRAL': 'EMBAKASI',
    'EMBAKASI NORTH': 'EMBAKASI',
    'RUARAKA': 'EMBAKASI',   # Assuming Ruaraka is part of Embakasi based on correction
}
# Apply corrections to 'Sub-County' column in facilities_df
facilities_df['Sub-County'] = facilities_df['Sub-County'].replace(corrections)

# Merge population_df and facilities_df on 'Sub-County'
merged_df = pd.merge(population_df, facilities_df, on='Sub-County', how='inner')


# Filter for Nairobi County since we are only interested in Nairobi
nairobi_df = merged_df[merged_df['County'] == 'Nairobi']
nairobi_df.info()

```

#### Exploratory Data Analysis
Removing non-significant columns

```python
# Drop columns with all NaN values
nairobi_df= nairobi_df.dropna(axis=1, how='all')

# Convert relevant columns to appropriate data types
nairobi_df['Beds'] = nairobi_df['Beds'].astype(int)
nairobi_df['Cots'] = nairobi_df['Cots'].astype(int)

#percentage of missing values in each column:

missing_percentages = nairobi_df.isnull().mean() * 100

#Identify columns with missing percentages greater than 60%
columns_to_drop = missing_percentages[missing_percentages > 60].index.tolist()

# non-significant columns based on my criteria
non_significant_columns = ['Cots']


# Drop identified columns from the DataFrame
df_cleaned = nairobi_df.drop(columns_to_drop, axis=1)

# Filter out rows where Operational Status is not 'Operational'
df_cleaned = df_cleaned[df_cleaned['Operational Status'] == 'Operational']

#Drop columns with high and low cardinality
df_cleaned.drop(columns=['Province', 'County','Constituency','Nearest Town','Town','Official Address','Description of Location','Post Code','In Charge','Job Title of in Charge', 'Operational Status','Location','Sub Location','Division']
, axis=1, inplace=True)

df_cleaned.info()
df_cleaned.tail(30)
```

####### Facility Density Calculation

Facility Density is a measure of the concentration of healthcare facilities within a specific area, typically expressed as the number of facilities per square kilometer. It is calculated using the formula:

\[ \text{Facility Density} = \frac{\text{Persons per Sq. Km}}{\text{Number of Facilities}} \]

```python
# Step 1: Calculate facility density per sub-county
facility_counts = df_cleaned.groupby('Sub-County')['Facility Code'].nunique()
population_density = df_cleaned.groupby('Sub-County')['Persons Per Sq. Km'].mean()

# Calculate facility density (facilities per sq. km)
facility_density = facility_counts / population_density

# Step 2: Sort facility density for better visualization
facility_density_sorted = facility_density.sort_values(ascending=False)

# Step 3: Plot bar graph of facility density per sub-county
plt.figure(figsize=(12, 6))
facility_density_sorted.plot(kind='bar', color='skyblue')
plt.title('Facility Density per Sub-County (Sorted)')
plt.xlabel('Sub-County')
plt.ylabel('Facility Density (Facilities per Sq. Km)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

Different sub-counties exhibit varying levels of facility density. For instance:

DAGORETTI has a relatively low facility density of 0.007043 facilities per square kilometer.
EMBAKASI and WESTLANDS have higher facility densities, with values around 0.023822 and 0.023 facilities per square kilometer, respectively.
MATHARE has the lowest facility density at 0.0002 facilities per square kilometer.

Facility density directly impacts how easily residents can access healthcare services:

High Facility Density: Areas like EMBAKASI and WESTLANDS, with higher facility densities, suggest that residents might have easier access to healthcare facilities within shorter distances. This can potentially lead to better healthcare service accessibility and shorter travel times for residents seeking medical care.
Low Facility Density: Conversely, sub-counties with lower facility densities, such as MATHARE, indicate that residents may face challenges in accessing healthcare services. Longer travel distances or extended travel times may be necessary to reach healthcare facilities, impacting accessibility.

```python

# Step 1: Count unique facilities per Sub-County
facility_counts = df_cleaned.groupby('Sub-County')['Facility Name'].nunique()

# Step 2: Merge Population Density (assuming it's already in df_cleaned)

# Step 3: Prepare X and y for regression
X = df_cleaned.groupby('Sub-County')['Persons Per Sq. Km'].first()  # Independent variable (Population Density)
y = facility_counts  # Dependent variable (Number of unique facilities)

# Ensure X and y have the same index
X = X.reindex(y.index)

# Step 4: Fit linear regression model
X = sm.add_constant(X)  # Add constant for intercept
model = sm.OLS(y, X).fit()

# Step 5: Print regression summary
print(model.summary())

# Interpretation
p_value_population_density = model.pvalues['Persons Per Sq. Km']

# Define significance level (alpha)
alpha = 0.05

# Step 6: Compare p-value to alpha
if p_value_population_density < alpha:
    print("\nReject the null hypothesis: Population density  significantly affects healthcare accessibility (measured by facility count).")
else:
    print("\nCannot reject the null hypothesis: Population density does not significantly affect healthcare accessibility (measured by facility count).")
```

```python

# Step 1: Group by Sub-County and calculate aggregate statistics
subcounty_stats = df_cleaned.groupby('Sub-County').agg({
    'Persons Per Sq. Km': 'mean',  # Average Population Density
    'Beds': 'sum'  # Total Beds in Healthcare Facilities
}).reset_index()

# Step 2: Prepare variables for regression
X = subcounty_stats['Persons Per Sq. Km']
y = subcounty_stats['Beds']

# Step 3: Fit linear regression model
X = sm.add_constant(X)  # Add constant for intercept
model = sm.OLS(y, X).fit()

# Step 4: Print regression summary
print(model.summary())

# Step 5: Interpret results
p_value_population_density = model.pvalues['Persons Per Sq. Km']
alpha = 0.05

if p_value_population_density < alpha:
    print("\nReject the null hypothesis: Population density  significantly affects bed accessibility.")
else:
    print("\nCannot reject the null hypothesis: Population density does not significantly affect bed accessibility.")
```
